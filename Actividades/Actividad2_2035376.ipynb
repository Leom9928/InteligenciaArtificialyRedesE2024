Desarrolla un algoritmo que por medio de una función, le pida al usuario una lista de N elementos y como resultado invierta el sentido de la lista.

```python
def invertir_lista(lista):
  lista_invertida = lista[::-1]
  return lista_invertida

# Solicitar al usuario el número de elementos de la lista
n = int(input("¿Cuántos elementos deseas ingresar en la lista? "))

# Solicitar al usuario los elementos de la lista
lista = []
for i in range(n):
  elemento = input(f"Ingresa el elemento {i+1}: ")
  lista.append(elemento)

# Invertir la lista
lista_invertida = invertir_lista(lista)

# Imprimir la lista invertida
print("La lista invertida es:", lista_invertida)
```

Desarrolla un algoritmo que le pregunte al usuario cuántos números de la serie de Fibonacci quiere generar y que al final los imprima.


```python
def fibonacci(n):
  if n <= 1:
    return [0, 1]
  else:
    fib_series = fibonacci(n - 1)
    fib_series.append(fib_series[-1] + fib_series[-2])
    return fib_series

# Solicitar al usuario cuántos números de la serie desea generar
n = int(input("¿Cuántos números de la serie de Fibonacci deseas generar? "))

# Generar los números de Fibonacci
fibonacci_series = fibonacci(n)

# Imprimir los números de Fibonacci
print("Los primeros", n, "números de la serie de Fibonacci son:", fibonacci_series)
```

Juego de adivinar el número: Crea un programa que genere un número aleatorio entre 1 y 100, y luego le pida al usuario que adivine el número. El programa debe proporcionar pistas al usuario si el número es mayor o menor que el número objetivo, y seguir solicitando un nuevo intento hasta que el usuario adivine correctamente.


```python
import random

# Generar un número aleatorio entre 1 y 100
numero_objetivo = random.randint(1, 100)

# Solicitar al usuario que adivine el número
adivinanza = -1
while adivinanza != numero_objetivo:
  adivinanza = int(input("Adivina el número (entre 1 y 100): "))

  if adivinanza > numero_objetivo:
    print("El número objetivo es menor.")
  elif adivinanza < numero_objetivo:
    print("El número objetivo es mayor.")

# El usuario ha adivinado correctamente
print("¡Felicidades! Has adivinado el número.")
```

Ejercicio de cálculo de números primos en un rango dado: Escribe un programa que solicite al usuario un rango de números y muestre todos los números primos dentro de ese rango.

```python
def es_primo(numero):
  if numero <= 1:
    return False
  for i in range(2, int(numero**0.5) + 1):
    if numero % i == 0:
      return False
  return True

# Solicitar al usuario el rango de números
inicio = int(input("¿Cuál es el número inicial del rango? "))
fin = int(input("¿Cuál es el número final del rango? "))

# Encontrar y mostrar los números primos en el rango
print("Números primos en el rango:", end=" ")
for numero in range(inicio, fin + 1):
  if es_primo(numero):
    print(numero, end=" ")

print()
```

Ejercicio de implementación de una calculadora básica utilizando funciones: Escribe un programa que solicite al usuario dos números y realice operaciones básicas como suma, resta, multiplicación y división utilizando funciones separadas para cada operación.

```python
def suma(a, b):
  return a + b

def resta(a, b):
  return a - b

def multiplicacion(a, b):
  return a * b

def division(a, b):
  if b == 0:
    raise ValueError("No se puede dividir por cero.")
  return a / b

# Solicitar al usuario los dos números
a = float(input("Ingresa el primer número: "))
b = float(input("Ingresa el segundo número: "))

# Mostrar el menú de operaciones
print("Selecciona la operación:")
print("1. Suma")
print("2. Resta")
print("3. Multiplicación")
print("4. División")

# Leer la opción del usuario
opcion = int(input("Ingresa el número de la operación: "))

# Realizar la operación seleccionada
if opcion == 1:
  resultado = suma(a, b)
elif opcion == 2:
  resultado = resta(a, b)
elif opcion == 3:
  resultado = multiplicacion(a, b)
elif opcion == 4:
  resultado = division(a, b)
else:
  print("Opción inválida.")
  exit()

# Imprimir el resultado
print("Resultado:", resultado)
```
